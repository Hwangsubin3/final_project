```python
import os
import urllib.request
import json 
import pandas as pd
import csv


nvMid=str(input("제품 코드를 입력해주세요: "))
score=str(input("리뷰 점수를 입력해주세요: "))

list=[]
for page in range(5):   #range() 안의 숫자는 페이지 반복 횟수,1페이지면 1
    url = "https://search.shopping.naver.com/api/review?nvMid="+nvMid+"&reviewType=ALL&sort=QUALITY&score="+score+"&isNeedAggregation=Y&isApplyFilter=Y&page="+str(page)+"&pageSize=20"
    request = urllib.request.Request(url)
    response = urllib.request.urlopen(request).read()
    
    text= json.loads(response)['reviews']
    review = pd.DataFrame(text)
    col=9
    reviews =review.iloc[:,col]
    list.append(reviews)
    
    #print(reviews)
    

    #txt파일로 저장

    #reviews.to_csv((nvMid+score)+'.txt',encoding='utf-8-sig')
    
    reviews_concat=pd.concat(list,ignore_index=True)
    print(reviews_concat)

    #csv 파일로 리뷰 저장
    reviews_concat.to_csv((nvMid+score)+'.csv',encoding='utf-8-sig')
    #reviews.to_csv('2페이지.csv',encoding='utf-8-sig')
    #reviews.to_csv('3페이지.csv',encoding='utf-8-sig')
```

```python
#텍스트 전처리하기
import re

f=open('SK매직_부정리뷰.txt','r',encoding='utf-8')
reviews=f.read()

def text_cleaning(word):
    text=re.compile('[^ ㄱ-ㅣ가-힣]+')
    text1=text.sub('',word)
    return text1
```

```python
new_review=text_cleaning(reviews)
```

```python
# 파일 저장
with open('SK매직_부정리뷰.txt', 'w') as f:
    f.write(new_review)
```

```python
from konlpy.tag import Okt
from collections import Counter
```

```python
f=open('SK매직_부정리뷰.txt','r',encoding='cp949')
reviews=f.read()

okt=Okt()
noun=okt.nouns(reviews)
stopwords={'식기세척기'}             #'식기세척기'는 워드클라우드 단어에서 제외되게 보여주기(빈도수多)
noun=[word for word in noun if word not in stopwords]
nouns = [n for n in noun if len(n) > 1]


count=Counter(nouns)

noun_list=count.most_common(50)
for i in noun_list:
    print(i)
```

```python
from wordcloud import WordCloud
#from wordcloud import STOPWORDS
my_path='C:\Windows\\Fonts\\malgunbd.ttf'
```

```python
# word cloud 객체 구성
wordcloud = WordCloud(font_path=my_path, 
                      background_color='white', 
                      width=1000, 
                      height=1000,
                      max_words=50,
                      max_font_size=300).generate_from_frequencies(dict(noun_list))

wordcloud.to_file('SK매직_부정리뷰.png')
```
